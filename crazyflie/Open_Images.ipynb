```python
import os
from PIL import Image
from pathlib import Path
from IPython.display import display

def open_design_images(design_name, base_directory=r"C:\Users\CrazyFlie\Documents\GitHub\CrazyConstruct\crazyflie"):
    folders = ['Claude_shapes', 'OpenAI_shapes', 'Gemini_Pro_1.5_shapes']
    images = []
    
    for folder in folders:
        file_path = Path(base_directory) / folder / (design_name + ".png")
        print(f"Looking for file at: {file_path}")  # Debugging line to check the full path
        
        if file_path.exists():
            image = Image.open(file_path)
            images.append(image)
        else:
            print(f"File not found: {file_path}")
    
    return images
```

```python
open_this = open_design_images(design_name = input())

# Display each image inlin
for img in open_this:
    display(img)
```
