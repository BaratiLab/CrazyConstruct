```python
# Specify the path to the .env file
env_path = Path(r'C:\Users\CrazyFlie\Documents\Nonuploadable_Git\.env')

# Load the .env file
load_dotenv(dotenv_path=env_path)

# Get the API key from environment variables
api_key = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-1.5-pro', generation_config={"response_mime_type": "application/json"})
```

```python
for design in design_requests:
    requested_design = design[0]
    current_coordinates = design[1]
    print("Current Coordinates: ", current_coordinates)
    prompt = functions.create_prompt(current_coordinates=current_coordinates, requested_design=requested_design)

    prompt = prompt
    
    # Send the request to OpenAI using the new ChatCompletion method
    response = model.generate_content(prompt)
    response_text = response.text
    #print(response_text)

    # Load JSON data
    data = json.loads(response_text)
    
    # Loop through each design and plot
    for entry in data['Design']:
        title = entry['Title']
        coordinates = entry['New Coordinates']
        print("New Coordiantes from LLM: ", coordinates)
        functions.plot_coordinates_2(title = title, list_of_coordinates=coordinates, figure_text=requested_design, list_of_current_coordinates = current_coordinates)
        #print(entry['Reasoning'])
        print("\n")
    time.sleep(60)
```
